## The environment: GridWorld

Our agent lives in a 4√ó4 grid world. Here's the layout:

```
+---+---+---+---+
| S |   |   |   |    S = Start (0,0)
+---+---+---+---+
|   |   | W | W |    W = Wall (can't enter)
+---+---+---+---+
|   |   |   | T |    T = Trap (reward: -5, episode ends)
+---+---+---+---+
| B |   |   | G |    B = Bait (reward: +3, episode ends)
+---+---+---+---+    G = Goal (reward: +10, episode ends)
```

**Actions:** 0 = Up, 1 = Right, 2 = Down, 3 = Left

**Rewards:** Each step costs -0.1 (to encourage efficiency). Reaching the goal gives +10, falling in the trap gives -5.

**Dynamics:** 15% of the time, the agent *slips* and moves in a random direction instead of the intended one.

> **Recall:** Using the lecture terminology, what are the *PEAS* of this environment? What are its *properties*?

Run the next cell to set up and visualize the environment.

````python
class GridWorld:
    """a 4x4 stochastic grid world with a goal, a trap, and a bait"""

    ACTIONS = {0: 'Up', 1: 'Right', 2: 'Down', 3: 'Left'}
    MOVES = {0: (-1, 0), 1: (0, 1), 2: (1, 0), 3: (0, -1)}

    def __init__(self, slip_prob=0.15):
        self.size = 4
        self.start = (0, 0)
        self.goal = (3, 3)          # +10 reward
        self.walls = {(1, 2), (1, 3)}
        self.traps = {(2, 3)}       # -5 reward, trap close to goal
        self.baits = {(3, 0): 3.0}  # +3 reward, tempting but suboptimal
        self.slip_prob = slip_prob   # 20% chance of random action
        self.agent_pos = self.start

    def reset(self):
        """reset agent to start. Returns initial state."""
        self.agent_pos = self.start
        return self.agent_pos

    def step(self, action):
        """take an action. Returns (new_state, reward, done).
        with slip_prob chance, the action is replaced by a random one."""
        # stochastic slip!
        if random.random() < self.slip_prob:
            action = random.randint(0, 3)

        dr, dc = self.MOVES[action]
        new_pos = (self.agent_pos[0] + dr, self.agent_pos[1] + dc)

        # only move if valid
        if (0 <= new_pos[0] < self.size and
            0 <= new_pos[1] < self.size and
            new_pos not in self.walls):
            self.agent_pos = new_pos

        # check outcome
        if self.agent_pos == self.goal:
            return self.agent_pos, 10.0, True
        elif self.agent_pos in self.traps:
            return self.agent_pos, -5.0, True
        elif self.agent_pos in self.baits:
            return self.agent_pos, self.baits[self.agent_pos], True
        else:
            return self.agent_pos, -0.1, False


# ===== Visualization & Episode Runner =====

def visualize_grid(env, trajectory=None, title="Grid World"):
    """draw the grid world. optionally show agent's path"""
    fig, ax = plt.subplots(1, 1, figsize=(5, 5))

    for i in range(env.size + 1):
        ax.axhline(i, color='black', linewidth=1)
        ax.axvline(i, color='black', linewidth=1)

    for r in range(env.size):
        for c in range(env.size):
            x, y = c + 0.5, env.size - 0.5 - r
            if (r, c) in env.walls:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#2c3e50', alpha=0.85))
                ax.text(x, y, 'W', ha='center', va='center', fontsize=18, fontweight='bold', color='white')
            elif (r, c) == env.goal:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#27ae60', alpha=0.25))
                ax.text(x, y, 'G\n+10', ha='center', va='center', fontsize=13, fontweight='bold', color='#27ae60')
            elif (r, c) in env.traps:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#e74c3c', alpha=0.20))
                ax.text(x, y, 'T\n-5', ha='center', va='center', fontsize=13, fontweight='bold', color='#e74c3c')
            elif (r, c) in env.baits:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#f39c12', alpha=0.25))
                bait_r = env.baits[(r, c)]
                ax.text(x, y, f'B\n+{bait_r:.0f}', ha='center', va='center', fontsize=13, fontweight='bold', color='#e67e22')
            elif (r, c) == env.start:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#3498db', alpha=0.15))
                ax.text(x, y, 'S', ha='center', va='center', fontsize=14, fontweight='bold', color='#3498db')

    if trajectory and len(trajectory) > 1:
        for i in range(len(trajectory)):
            r, c = trajectory[i]
            x, y = c + 0.5, env.size - 0.5 - r
            if i == 0:
                ax.plot(x, y, 'o', color='#3498db', markersize=12, zorder=5)
            if i > 0:
                pr, pc = trajectory[i-1]
                px, py = pc + 0.5, env.size - 0.5 - pr
                jx = random.uniform(-0.05, 0.05)
                jy = random.uniform(-0.05, 0.05)
                ax.annotate('', xy=(x+jx, y+jy), xytext=(px+jx, py+jy),
                           arrowprops=dict(arrowstyle='->', color='#e67e22', lw=2, alpha=0.6))
            if i == len(trajectory) - 1:
                ax.plot(x, y, 's', color='#e67e22', markersize=10, zorder=5)

    ax.set_xlim(0, env.size); ax.set_ylim(0, env.size)
    ax.set_aspect('equal'); ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])
    plt.tight_layout(); plt.show()


def run_episode(env, agent, max_steps=50, train=False):
    """run one episode. returns (total_reward, trajectory)"""
    state = env.reset()
    if hasattr(agent, 'reset'):
        agent.reset()

    total_reward = 0
    trajectory = [state]

    for _ in range(max_steps):
        action = agent.choose_action(state)
        next_state, reward, done = env.step(action)

        if train and hasattr(agent, 'learn'):
            agent.learn(state, action, reward, next_state, done)

        total_reward += reward
        trajectory.append(next_state)
        state = next_state

        if done:
            break

    return total_reward, trajectory


print("‚úÖ Environment and helpers ready!")

env = GridWorld()
visualize_grid(env)
````

## 1. Simple Reflex Agent (Pre-Built)

This is the simplest agent from the lecture: it follows *fixed rules* based only on its *current position*. It has no memory, planning, or learning.

Its rules are: go right if possible; else go down; else go randomly.

Run the cells below to see how it performs.

````python
class SimpleReflexAgent:
    """Fixed if-then rules, no memory"""

    def __init__(self):
        self.name = "Simple Reflex"

    def choose_action(self, state):
        row, col = state
        if col < 3:       # can go right?
            return 1       #   -> go right
        elif row < 3:      # can go down?
            return 2       #   -> go down
        else:              # stuck?
            return random.choice([0, 1, 2, 3])  # -> random

    def reset(self):
        pass

# run it!
env = GridWorld()
simple_agent = SimpleReflexAgent()
results = []

# since env is stochastic, even a rule-based agent can see different outcomes
# => we'll run it a few times
for i in range(10):
    random.seed(i * 42)
    reward, traj = run_episode(env, simple_agent)
    results.append((reward, traj))

print(f"Agent: {simple_agent.name}")
print(f"\n10-run summary:")
for i, (r, t) in enumerate(results):
    final = t[-1]
    if final == env.goal: outcome = "GOAL ‚úì"
    elif final in env.traps: outcome = "TRAP ‚úó"
    elif final in env.baits: outcome = "BAIT ~"
    elif len(t) > 49: outcome = "STUCK ‚úó"
    else: outcome = "???"
    print(f"  Run {i+1:>2d}: reward = {r:>5.1f}  ‚Üí  {outcome}")

# Show a typical run
reward, trajectory = results[0]
visualize_grid(env, trajectory, title=f"Simple Reflex Agent (reward: {reward:.1f})")
````

### => Reflex agent gets stuck üöß

The simple reflex agent walks to the right, and gets stuck at (0,3) trying to run downwards into the wall. It doesn't see the wall (only its own position), and it doesn't remember getting stuck there in the past.

## Task 2: Q-Learning Agent that improves with experience (~10 min)

### How Q-learning works

The agent maintains a table $Q(s, a)$, a "quality score" for each state-action pair $(s,a)$. Higher Q means "this action in this state tends to lead to good outcomes." The agent learns/improves by exploring different states & actions and updating the Q-table with the rewards it sees.

**Example:** once the agent walks into the trap, it'll lower the value of "at state [left of trap], go right", and thus will prefer to e.g. go down instead (higher value). Once it's also seen the goal state, it'll increase the value of "at state [left of trap], go down", because in the episodes where it's done that, this led to a high reward in the end.

#### Learning rule `QLearningAgent.learn(s,a,r,s')`:
After taking action $a$ in state $s$, getting reward $r$, and landing in state $s'$:

$$Q(s, a) \leftarrow Q(s, a) + \alpha \cdot \Big[ r + \gamma \cdot \max_{a'} Q(s', a') - Q(s, a) \Big]$$

Where:
- **Œ±** (learning rate): how fast we update (0.1 = cautious, 0.9 = aggressive)
- **Œ≥** (discount factor): how much we value future rewards (0.95 = future matters a lot)

#### Action selection `QLearningAgent.choose_action(s,a)`:
We'll use what's called *"epsilon-greedy"*:
- With probability **Œµ**: explore = pick random action
- With probability **1 - Œµ**: exploit = pick best known ("greedy") action

### üîó Connection to the lecture

| Agent component | Q-Learning equivalent |
|---|---|
| **Performance Element** | Choosing actions using Q-values |
| **Critic** | The reward signal from the environment |
| **Learning Element** | The Q-value update rule |
| **Problem Generator** | Epsilon exploration (trying random actions to discover new things) |

> üí° **Fun fact:** Q-learning is a direct ancestor of the techniques used to fine-tune modern LLMs, like *Reinforcement Learning from Human Feedback (RLHF)*!

Now let's train the Q-learning agent over many episodes and watch it get smarter.

Run the training cell below. You should see the reward curve *go up* as the agent learns!

````python
# Train the Q-learning agent
env = GridWorld()
q_agent = QLearningAgent(learning_rate=0.2, discount_factor=0.95, epsilon=0.3)

n_episodes = 800
rewards_history = []

for ep in range(n_episodes):
    reward, _ = run_episode(env, q_agent, max_steps=50, train=True)
    rewards_history.append(reward)

# Plot the learning curve
fig, axes = plt.subplots(1, 2, figsize=(14, 4.5))

# Raw rewards
axes[0].plot(rewards_history, alpha=0.3, color='#3498db', linewidth=0.8)
axes[0].set_title('Reward per Episode (Raw)', fontweight='bold')
axes[0].set_xlabel('Episode'); axes[0].set_ylabel('Total Reward')
axes[0].axhline(y=10, color='#27ae60', linestyle='--', alpha=0.5, label='Goal (+10)')
axes[0].axhline(y=3, color='#e67e22', linestyle=':', alpha=0.5, label='Bait (+3)')
axes[0].axhline(y=-5, color='#e74c3c', linestyle='--', alpha=0.5, label='Trap (-5)')
axes[0].legend(fontsize=9); axes[0].grid(True, alpha=0.3)

# Smoothed
window = 40
smoothed = np.convolve(rewards_history, np.ones(window)/window, mode='valid')
axes[1].plot(smoothed, color='#3498db', linewidth=2)
axes[1].set_title(f'Reward per Episode (Smoothed)', fontweight='bold')
axes[1].set_xlabel('Episode'); axes[1].set_ylabel('Average Reward')
axes[1].axhline(y=10, color='#27ae60', linestyle='--', alpha=0.5, label='Goal (+10)')
axes[1].axhline(y=3, color='#e67e22', linestyle=':', alpha=0.5, label='Bait (+3)')
axes[1].legend(fontsize=9); axes[1].grid(True, alpha=0.3)

plt.tight_layout(); plt.show()

# Summary
last_100 = rewards_history[-100:]
goals = sum(1 for r in last_100 if r > 8)
baits = sum(1 for r in last_100 if 1 < r < 5)
traps = sum(1 for r in last_100 if r < -3)
print(f"\nüìä Training summary ({n_episodes} episodes):")
print(f"   First 50 episodes ‚Äî avg reward: {np.mean(rewards_history[:50]):.1f}")
print(f"   Last 100 episodes ‚Äî avg reward: {np.mean(last_100):.1f}")
print(f"   Last 100: {goals} goals, {baits} baits, {traps} traps")
````

### Let's watch the trained agent in action üëÄ

````python
# Wwtch the trained agent (with no exploration ‚Äî pure exploitation)
env = GridWorld()
old_eps = q_agent.epsilon
q_agent.epsilon = 0  # Turn off exploration

# run a few times to show a successful run (stochastic env means some episodes slip)
for attempt in range(10):
    random.seed(attempt + 1)
    reward, trajectory = run_episode(env, q_agent)
    if trajectory[-1] == env.goal:
        break

q_agent.epsilon = old_eps

print(f"Trained agent: reward = {reward:.1f}, steps = {len(trajectory)-1}")
print(f"Path: {' ‚Üí '.join(str(p) for p in trajectory)}")
visualize_grid(env, trajectory, title=f"Trained Q-Learning Agent (reward: {reward:.1f})")
````

## Experiment: Hyperparameter Tuning

The two key parameters of Q-learning are **Œµ** (exploration rate) and **Œ±** (learning rate). Both involve real tradeoffs ‚Äî let's see them in action.


### Experiment A: Exploration Rate Œµ

We fix Œ± = 0.5 and sweep over different Œµ values.

> ü§î **Prediction before you run this:** With Œµ = 0.001 (almost no exploration), the agent will quickly find *something*. But will it be the bait or the goal?

````python
# Experiment A: Effect of exploration rate (epsilon)
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

test_epsilons = [0.001, 0.25, 0.5, 0.8]
colors = ['#e74c3c', '#27ae60', '#3498db', '#9b59b6']
n_seeds = 20

for eps, col in zip(test_epsilons, colors):
    all_rewards = []
    for seed in range(n_seeds):
        random.seed(seed * 77); np.random.seed(seed * 77)
        agent = QLearningAgent(learning_rate=0.5, discount_factor=0.95, epsilon=eps)
        rewards = [run_episode(GridWorld(), agent, max_steps=50, train=True)[0]
                   for _ in range(1000)]
        all_rewards.append(rewards)
    avg_rewards = np.mean(all_rewards, axis=0)
    smoothed = np.convolve(avg_rewards, np.ones(40)/40, mode='valid')
    ax.plot(smoothed, label=f'Œµ = {eps}', color=col, linewidth=2.5)

ax.axhline(y=3.0, color='gray', linestyle=':', alpha=0.4, label='Bait level (+3)')
ax.set_title('Effect of Exploration Rate Œµ  (Œ± = 0.5 fixed)', fontsize=14, fontweight='bold')
ax.set_xlabel('Episode', fontsize=12); ax.set_ylabel('Average Reward (smoothed)', fontsize=12)
ax.legend(fontsize=11); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
````

### üí° What you should see:

- **Œµ = 0.001:** The agent quickly finds the bait (+3) and *stays there forever*. With barely any exploration, it never discovers the much better goal. This is the *local optimum trap*. it settles for "good enough" because it never tried anything new.

- **Œµ = 0.25:** The sweet spot. Enough exploration to discover the goal, enough exploitation to converge.

- **Œµ = 0.5:** Finds goal quicker, but often wanders into the trap due to higher randomness, noisier convergence.

- **Œµ = 0.8:** Almost random behavior! Can't consistently act on what it learned.

> üåü **Connection:** This is exactly the *exploration-exploitation dilemma*. It shows up everywhere from how you pick restaurants (safe favorite vs. try something new?) to how tech companies A/B test features.

### Experiment B: Learning Rate Œ±

Now we fix Œµ = 0.3 and sweep over different learning rates.

````python
# Experiment B: Effect of learning rate (alpha)
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

test_lrs = [0.05, 0.2, 0.5, 0.9]
colors = ['#e74c3c', '#27ae60', '#3498db', '#9b59b6']

for lr, col in zip(test_lrs, colors):
    all_rewards = []
    for seed in range(n_seeds):
        random.seed(seed * 77); np.random.seed(seed * 77)
        agent = QLearningAgent(learning_rate=lr, discount_factor=0.95, epsilon=0.3)
        rewards = [run_episode(GridWorld(), agent, max_steps=50, train=True)[0]
                   for _ in range(1000)]
        all_rewards.append(rewards)
    avg_rewards = np.mean(all_rewards, axis=0)
    smoothed = np.convolve(avg_rewards, np.ones(40)/40, mode='valid')
    ax.plot(smoothed, label=f'Œ± = {lr}', color=col, linewidth=2.5)

ax.axhline(y=3.0, color='gray', linestyle=':', alpha=0.4, label='Bait level (+3)')
ax.set_title('Effect of Learning Rate Œ±  (Œµ = 0.3 fixed)', fontsize=14, fontweight='bold')
ax.set_xlabel('Episode', fontsize=12); ax.set_ylabel('Average Reward (smoothed)', fontsize=12)
ax.legend(fontsize=11); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
````

### üí° What you should see:

- **Œ± = 0.05:** Stuck at the bait level for the entire training! The updates are so tiny that Q-values barely change. The agent finds the bait first and can never update fast enough to learn that the goal is better.

- **Œ± = 0.2:** Good convergence. Learns to prefer the goal over the bait.

- **Œ± = 0.5:** Also good. Faster initial learning.

- **Œ± = 0.9:** Fast at the start, but notice the **noisier curve**. In our stochastic environment (15% slip), a high learning rate means each lucky or unlucky episode massively overwrites the previous Q-value estimate. The values keep bouncing around.

=> A higher LR isn't always better. In a stochastic environments you can randomly seeing local optimum trajectories (i.e., hitting the bait), and with a high LR you run the danger of "locking in" on them too early, before you've seen the true goal state.

## Visualizing the Learned Policy

If you have time, run this cell to see *what the agent learned*, shown as arrows indicating the best action at each cell, with colors showing how valuable it thinks each cell is.

````python
def visualize_policy(env, agent, title="Learned Policy"):
    """Show the agent's learned policy as arrows on the grid."""
    fig, ax = plt.subplots(1, 1, figsize=(6, 6))
    arrow_dx = {0: 0, 1: 0.25, 2: 0, 3: -0.25}
    arrow_dy = {0: 0.25, 1: 0, 2: -0.25, 3: 0}

    for i in range(env.size + 1):
        ax.axhline(i, color='black', linewidth=1)
        ax.axvline(i, color='black', linewidth=1)

    for r in range(env.size):
        for c in range(env.size):
            x, y = c + 0.5, env.size - 0.5 - r
            if (r, c) in env.walls:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#2c3e50', alpha=0.85))
                ax.text(x, y, 'W', ha='center', va='center', fontsize=16, fontweight='bold', color='white')
            elif (r, c) == env.goal:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#27ae60', alpha=0.4))
                ax.text(x, y, 'G', ha='center', va='center', fontsize=18, fontweight='bold', color='#27ae60')
            elif (r, c) in env.traps:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#e74c3c', alpha=0.3))
                ax.text(x, y, 'T', ha='center', va='center', fontsize=18, fontweight='bold', color='#e74c3c')
            elif (r, c) in env.baits:
                ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1, color='#f39c12', alpha=0.3))
                ax.text(x, y, 'B', ha='center', va='center', fontsize=18, fontweight='bold', color='#e67e22')
            else:
                q_vals = [agent.get_q((r, c), a) for a in range(4)]
                best_a = int(np.argmax(q_vals))
                max_q = max(q_vals)
                if max_q != 0 or min(q_vals) != 0:
                    norm_q = np.clip((max_q + 5) / 15, 0, 1)
                    ax.add_patch(plt.Rectangle((c, env.size-1-r), 1, 1,
                                color=(0.2, 0.7, 0.3, norm_q * 0.5)))
                    ax.annotate('', xy=(x + arrow_dx[best_a], y + arrow_dy[best_a]),
                               xytext=(x, y),
                               arrowprops=dict(arrowstyle='->', color='#2c3e50', lw=2.5))
                    ax.text(x, y - 0.37, f'{max_q:.1f}', ha='center', va='center',
                           fontsize=7, color='#555')

    ax.set_xlim(0, env.size); ax.set_ylim(0, env.size)
    ax.set_aspect('equal')
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xticks([]); ax.set_yticks([])
    plt.tight_layout(); plt.show()

visualize_policy(env, q_agent, title="Q-Learning Agent ‚Äî Learned Policy")
print("Arrows: best action per cell. Greener = higher Q-value.")
print("Notice: the agent has learned to avoid the trap, ignore the bait,")
print("and navigate toward the goal!")
````

## üìä Final Comparison: All Three Agents

````python
# Head-to-head comparison
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

agents = [
    ("Simple Reflex", lambda: SimpleReflexAgent()),
    ("Model-Based", lambda: ModelBasedReflexAgent()),
    ("Q-Learning", lambda: QLearningAgent(learning_rate=0.5, discount_factor=0.95, epsilon=0.25)),
]
colors = ['#e74c3c', '#e67e22', '#3498db']

for (name, make_agent), color in zip(agents, colors):
    all_rewards = []
    for seed in range(5):
        random.seed(seed * 77); np.random.seed(seed * 77)
        agent = make_agent()
        rewards = [run_episode(GridWorld(), agent, max_steps=50, train=True)[0]
                   for _ in range(800)]
        all_rewards.append(rewards)
    avg = np.mean(all_rewards, axis=0)
    smoothed = np.convolve(avg, np.ones(40)/40, mode='valid')
    ax.plot(smoothed, label=name, color=color, linewidth=2.5)

ax.axhline(y=3.0, color='gray', linestyle=':', alpha=0.4)
ax.set_title('Agent Comparison: Reflex ‚Üí Model-Based ‚Üí Learning', fontsize=14, fontweight='bold')
ax.set_xlabel('Episode', fontsize=12); ax.set_ylabel('Average Reward (smoothed)', fontsize=12)
ax.legend(fontsize=12); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
````

### Key takeaways:
- Simple reflex: stuck on the wall forever (flat line at the bottom)
- Model-based: navigates walls, finds bait ‚Äî better but plateaus
- Q-learning: starts like model-based, then KEEPS IMPROVING
